```python
import sys
import numpy as np
import cv2
import pandas as pd
```

```python
img = cv2.imread('./king_face.png')

if img is None:
    print('image read failed')
    sys.exit()
    
## Caffe model
model = './res10_300x300_ssd_iter_140000_fp16.caffemodel'
config = './deploy.prototxt.txt'

## tensorflow
model = './opencv_face_detector_uint8.pb'
config = './opencv_face_detector.pbtxt.txt'

## model import
face_net = cv2.dnn.readNet(model, config)

if face_net.empty():
    print('model read failed')
    sys.exit()
    
blob = cv2.dnn.blobFromImage(img, 1, (300,300), (104,177,123), 
                            swapRB = False)
face_net.setInput(blob)
out = face_net.forward()

labels = ['img_id', 'is_face', 'confidence', 'x1', 'y1', 'x2', 'y2']
out_df = pd.DataFrame(out[0][0], columns = labels)

# out_df
# print(out)

detect = out[0,0, :, :]
h, w = img.shape[:2]

for i in range(detect.shape[0]):
    confidence = detect[i,2]
    
    if confidence > 0.85:
        x1 = int(detect[i,3]*w)
        y1 = int(detect[i,4]*h)
        x2 = int(detect[i,5]*w)
        y2 = int(detect[i,6]*h)
        
        cv2.rectangle(img, (x1,y1), (x2,y2), (0,0,255), 1, cv2.LINE_AA)
        text = f'Face,{confidence*100:.2f}%'
        cv2.putText(img,text, (x1,y1-3), cv2.FONT_HERSHEY_COMPLEX, 0.5,
                   (0,0,255), 1, cv2.LINE_AA)
        
cv2.imshow('image',img)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
cap = cv2.VideoCapture(0)

if not cap.isOpened():
    print('camera read failed')
    sys.exit()

w = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
h = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
fps = int(cap.get(cv2.CAP_PROP_FPS))

while True:
    ret, frame = cap.read()
    
    if not ret:
        print('frame read failed')
        break
    
## Caffe model
    model = './res10_300x300_ssd_iter_140000_fp16.caffemodel'
    config = './deploy.prototxt.txt'

    ## tensorflow
    model = './opencv_face_detector_uint8.pb'
    config = './opencv_face_detector.pbtxt.txt'

    ## model import
    face_net = cv2.dnn.readNet(model, config)

    if face_net.empty():
        print('model read failed')
        sys.exit()
    
    blob = cv2.dnn.blobFromImage(frame, 1, (300,300), (104,177,123), 
                                swapRB = False)
    face_net.setInput(blob)
    out = face_net.forward()

    labels = ['img_id', 'is_face', 'confidence', 'x1', 'y1', 'x2', 'y2']
    out_df = pd.DataFrame(out[0][0], columns = labels)

# out_df
# print(out)

    detect = out[0,0, :, :]

    for i in range(detect.shape[0]):
        confidence = detect[i,2]

        if confidence > 0.5:
            x1 = int(detect[i,3]*w)
            y1 = int(detect[i,4]*h)
            x2 = int(detect[i,5]*w)
            y2 = int(detect[i,6]*h)

            cv2.rectangle(frame, (x1,y1), (x2,y2), (0,0,255), 1, cv2.LINE_AA)
            text = f'Face,{confidence*100:.2f}%'
            cv2.putText(frame,text, (x1,y1-3), cv2.FONT_HERSHEY_COMPLEX, 0.5,
                       (0,0,255), 1, cv2.LINE_AA)
        
    cv2.imshow('frame',frame)
    if cv2.waitKey(50) == 27:
        break
    
cap.release()
cv2.destroyAllWindows()
```
