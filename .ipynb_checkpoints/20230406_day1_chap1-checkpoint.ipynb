```python
image = cv2.imread('./Ch_1_fig/fig/my_images/KakaoTalk_20230406_151400751.jpg', cv2.IMREAD_COLOR)
# print(image.shape)
h, w = image.shape[:2]

img_resize=cv2.resize(src=image, dsize=(w//2,h//2), interpolation=cv2.INTER_AREA)

# cv2.imwrite('./Ch_1_fig/fig/img_resize.png', img_resize)
# filename: 저장할 파일 이름
# img: 저장할 자료 (numpy.ndarray)
# params: 파일저장 옵션
# retval: 성공 True, 실패 False

if image is None:
    print('image read failed')
    sys.exit()

cv2.namedWindow('image')
cv2.imshow('image',img_resize)

cv2.waitKey()
cv2.destroyAllWindows()
```

```python
img = cv2.imread('./Ch_1_fig/fig/puppy.bmp')

if img is None:
    print('image read failed')
    sys.exit()
    
cv2.namedWindow('image', cv2.WINDOW_AUTOSIZE)
cv2.imshow('image',img)

while True:
    if cv2.waitKey() == 27 or ord('q'):
        break

cv2.destroyAllWindows()
```

```python
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('./Ch_1_fig/fig/puppy.bmp')
# print(img.shape) # BGR
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
img_gray = cv2.cvtColor(img, cv2.IMREAD_GRAYSCALE)

if img is None:
    print('image read failed')
    sys.exit()

fig, ax = plt.subplots(1,3)
ax[0].imshow(img), ax[0].set_title('BGR'), ax[0].axis('off')
ax[1].imshow(img_rgb), ax[1].set_title('RGB'), ax[1].axis('off')
ax[2].imshow(img_gray), ax[2].set_title('GRAY'), ax[2].axis('off')
plt.show()

# plt.figure(figsize = (5,5))
# plt.imshow(img) # RGB
# plt.axis('off')
# plt.show()

cv2.namedWindow('image', cv2.WINDOW_AUTOSIZE)
cv2.imshow('image',img)

while True:
    if cv2.waitKey() == 27 or ord('q'):
        break
        
cv2.destroyAllWindows()
```

```python
import os
```

```python
img_lists = os.listdir('./Ch_1_fig/fig/images/')
img_paths =[]
for i in img_lists:
    img_path = './Ch_1_fig/fig/images/' + i
    img_paths.append(img_path)
    
    cv2.namedWindow('image', cv2.WINDOW_NORMAL)
    cv2.setWindowProperty('image', cv2.WND_PROP_FULLSCREEN,
                         cv2.WINDOW_FULLSCREEN)
idx = 0
while True:
    img = cv2.imread(img_paths[idx])
    
    if img is None:
        print('image read failed')
        continue
        
    cv2.imshow('image',img)
    
    if cv2.waitKey() == 27 or ord('q'):
        break
    
    idx += 1
    
    if idx >= len(img_paths):
        idx = 0
cv2.destroyAllWindows()
```
