```python
import sys
import os
import matplotlib.pyplot as plt
import numpy as np
import cv2
```

```python
img = cv2.imread('./Ch_1_fig/fig/puppy.bmp', cv2.IMREAD_COLOR)
# print(type(img))
# print(img.dtype)
# print(img.shape)
h, w = img.shape[:2]
img_resize = cv2.resize(img,(w//2,h//2))
cv2.imwrite('./Ch_1_fig/fig/my_images/resize.jpg',img_resize)

if img is None:
    print('image read failed')
    sys.exit()
    
cv2.namedWindow('image',cv2.WINDOW_AUTOSIZE)
cv2.imshow('image', img_resize)

while True:
    key = cv2.waitKey()
    if key==27 or ord('q'):
        break
        
cv2.destroyAllWindows()
```

```python
img_BGR = cv2.imread('./Ch_1_fig/fig/my_images/KakaoTalk_20230406_151400751.jpg', cv2.IMREAD_REDUCED_COLOR_2)
img_RGB = cv2.cvtColor(img_BGR, cv2.COLOR_BGR2RGB)
img_GRAY = cv2.cvtColor(img_BGR, cv2.COLOR_BGR2GRAY)

if img_BGR is None:
    print('image read failed')
    sys.exit()
    
fig, ax = plt.subplots(1, 3, figsize = (12,6))

ax[0].imshow(img_BGR), ax[0].set_title('img_BGR'), ax[0].axis('off')
ax[1].imshow(img_RGB), ax[1].set_title('img_RGB'), ax[1].axis('off')
ax[2].imshow(img_GRAY), ax[2].set_title('img_GRAY'), ax[2].axis('off')

plt.show()
```
